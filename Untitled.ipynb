```python
ride_time_df.index = ride_time_df['ride_id']
```

```python
ride_time_df.drop(labels = 'ride_id',inplace = True,axis =1)
```

```python
rides = rides_df.join(ride_time_df,on = 'ride_id',lsuffix = 'l',rsuffix = 'r')
```
